```python
import pandas as pd
```

```python
individuals_data_frame = pd.read_csv("./individuals.csv")
individuals_data_frame = individuals_data_frame.sort_values(
    by=["country", "branch_no", "department_no", "department", "head_no"]
)
individuals_data_frame["department_no"] = individuals_data_frame[
    "department_no"
].astype(int)
# individuals_data_frame['head_no'] = individuals_data_frame['head_no'].astype(int)
individuals_data_frame.to_csv("./individuals.csv", index=False)
individuals_data_frame
```

## Individuals

```python
unique_individuals_data_frame = individuals_data_frame[["name", "head", "department"]]
unique_individuals_data_frame = unique_individuals_data_frame.drop_duplicates(
    ["name", "head", "department"]
)
unique_individuals_data_frame = unique_individuals_data_frame.sort_values(by=["name"])
unique_individuals_data_frame
```

## Legistrative

### Speakers of the House of Representatives

```python
legistrative_data_frame = individuals_data_frame[
    individuals_data_frame["branch"] == "legistrative"
]
legistrative_data_frame
```

```python
speakers_data_frame = legistrative_data_frame[
    legistrative_data_frame["head"] == "Speaker"
]
```

```python
speakers = speakers_data_frame["name"].unique()
print(len(speakers))
speakers
```

```python
states = speakers_data_frame["state"].unique()
print(len(states))
states
```

## Executive

### Departments

```python
department_data_frame = individuals_data_frame[
    individuals_data_frame["department_no"] != 0
]
department_data_frame = department_data_frame[["department_no", "department", "head"]]
department_data_frame = department_data_frame.drop_duplicates(
    ["department_no", "department", "head"]
)
department_data_frame
```

### Secretaries of State

```python
state_data_frame = individuals_data_frame[
    individuals_data_frame["department"] == "Department of State"
]
state_data_frame
```

```python
foreign_borned_state_data_frame = state_data_frame[
    state_data_frame["birth_country"] != "United States"
]
foreign_borned_state_data_frame
```
