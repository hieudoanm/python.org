```python
import pandas as pd
```

```python
governors_data_frame = pd.read_csv("./governors.csv")
governors_data_frame["state_admission_date"] = pd.to_datetime(
    governors_data_frame["state_admission_date"]
)
governors_data_frame
```

## States

```python
states_data_frame = governors_data_frame[governors_data_frame["state_level"] == "state"]
columns = [
    "region",
    "subregion",
    "state",
    "state_admission_date",
    "state_capital",
    "state_largest",
]
states_data_frame = states_data_frame[columns]
states_data_frame = states_data_frame.sort_values(by=["state_admission_date"])
states_data_frame = states_data_frame.drop_duplicates(columns)
print(len(states_data_frame))
states_data_frame
```
