# Passport

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
from tqdm.notebook import tqdm as tdqm_notebook
```

```python
response = requests.get("https://restcountries.com/v3/all")
countries = response.json()
```

countries_data_frame = pd.DataFrame(countries)
countries_data_frame = countries_data_frame.sort_values(by=["cca3"])
countries_data_frame

```python
visas: list[dict] = []

for country in tdqm_notebook(countries):
    independent = country.get("independent", False)
    if not independent:
        continue
    country_common_name: str = country.get("name", {}).get("common", "")
    country_path: str = "-".join(country_common_name.split(" ")).lower()
    try:
        print(country_path)
        URL = f"https://www.passportindex.org/passport/{country_path}"
        response = requests.get(URL)
        html = response.text
        soup = BeautifulSoup(html, "html.parser")
        rows = soup.find_all("table")[0].find("tbody").find_all("tr")
        rowsList = list(rows)
        for row in rows:
            cells = row.find_all("td")
            cellsList = list(cells)
            country_text = cellsList[0].getText().strip()
            requirement_text = cellsList[1].getText().strip()
            visa: dict = {
                "to_country": country_common_name,
                "from_country": country_text,
                "requirement": requirement_text,
            }
            visas.append(visa)
    except Exception as exception:
        print(country_path, exception)

visas_data_frame = pd.DataFrame(visas)
visas_data_frame = visas_data_frame.sort_values(
    by=["to_country", "from_country", "requirement"]
)
visas_data_frame.to_csv("./visas.csv", index=False, header=True)
```
