```python
from pandas import DataFrame, read_csv


dataset = read_csv("./full_dataset.csv")
dataset
```

```python
languages = list(dataset["Language"].unique())
languages.sort()
languages
```

```python
for index, language in enumerate(languages):
    language_dataset: DataFrame = dataset[dataset["Language"] == language]
    language_dataset.to_csv(f"./training/{language}.csv", index=False)
    print(index + 1, language, language_dataset["Text"].count())
```
