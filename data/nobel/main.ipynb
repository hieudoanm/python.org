# Nobel Laureates

```python
import pandas
```

```python
laureates_data_frame = pandas.read_csv("./laureates.csv")
laureates_data_frame = laureates_data_frame.sort_values(
    by=["Category", "Year"], ignore_index=True
)
laureates_data_frame.to_csv("./laureates.csv", index=False, header=True)
laureates_data_frame.columns = map(
    lambda item: "_".join(item.lower().split(" ")), laureates_data_frame.columns
)
laureates_data_frame = laureates_data_frame.sort_values(
    by=["category", "year"], ignore_index=True
)
laureates_data_frame.to_csv(
    "./transformed/laureates.csv", index_label="id", header=True, na_rep="N/A"
)
laureates_data_frame
```
