```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
def generate_results(x):
  return 2 * x + 1
sample_input = list(range(20))
sample_output = list(map(generate_results, sample_input))
print(sample_input)
print(sample_output)
```

```python
plt.scatter(sample_input, sample_output)
plt.xlim(0, 20)
plt.ylim(0, 40)
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

```python
layer_0 = keras.layers.Dense(units=1, activation="relu", input_shape=[1])
model = keras.Sequential([
  layer_0
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(1), loss="mean_squared_error")
```

```python
trained_model = model.fit(sample_input, sample_output, epochs=500)
```

```python
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(trained_model.history['loss'])
plt.show()
```

```python
test_number = 60
predicted_numbers = model.predict([test_number])
print("The Predicted Number: " + str(predicted_numbers))
result_number = test_number * 2 + 1
print("The Result Number:" + str(result_number))
difference = result_number - predicted_numbers
print("Prediction error:" + str(difference))
```
