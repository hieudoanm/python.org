```python
import json

with open("./all/openings.json", "r") as json_file:
    openings : list[dict[str]] = json.load(json_file)

list_of_moves = list(
    map(
        lambda o: [value for index, value in enumerate(o.get("pgn", "").split(" ")) if index % 3 != 0],
        openings,
    )
)

list_of_moves.sort()

list_of_moves
```

```python
def build_tree(data):
    tree = {}
    for path in data:
        current = tree
        for key in path:
            current = current.setdefault(key, {})  # Move deeper in the tree
            print(key)
    return tree

tree = build_tree(list_of_moves)

tree
```

```python
from io import TextIOWrapper

with open("./all/tree.json", "w", encoding="utf-8") as tree_json_file:
    json.dump(tree, tree_json_file, ensure_ascii=False, indent=2)
```
